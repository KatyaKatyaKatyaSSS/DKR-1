# **Задача 1.**

**Вычисление математического ожидания, дисперсии и среднеквадратичного отклонения для варианта 1.**

Данные для варианта 1 взяты из таблицы:

*Значения случайной величины X:* 10, 13, 17, 20, 25


*Их вероятности:* 0.04, 0.3, 0.17, 0.15, 0.05

```python
import numpy as np

# Данные для варианта 1
X = np.array([10, 13, 17, 20, 25])  # значения случайной величины
P = np.array([0.04, 0.3, 0.17, 0.15, 0.05])  # вероятности из таблицы

# Вычисляем недостающую вероятность
missing_probability = 1 - P.sum()

# Добавляем эту вероятность равномерно к элементам массива
P += missing_probability / len(P)

# Проверяем, что сумма вероятностей теперь равна 1
assert np.isclose(P.sum(), 1), "Ошибка: сумма вероятностей должна быть 1."

# Математическое ожидание
expected_value = np.sum(X * P)

# Дисперсия
variance = np.sum(P * (X - expected_value) ** 2)

# Среднеквадратичное отклонение
std_deviation = np.sqrt(variance)

# Вывод результатов
print(f"Математическое ожидание (E[X]): {expected_value:.2f}")
print(f"Дисперсия (D[X]): {variance:.2f}")
print(f"Среднеквадратичное отклонение (σ): {std_deviation:.2f}")
```

# **Задача 3**

Цифра измерительного прибора округляется до ближайшего деления (0.3), найти вероятность ошибки:

**а) меньше 0.4**

**б) больше 0.5**

Используем нормальное распределение, предполагая, что округление соответствует погрешности.

```python
from scipy.stats import norm

# Среднее значение и стандартное отклонение
mu = 0.4
sigma = 0.05

# Вероятности
p_less_04 = norm.cdf(0.4, mu, sigma)
p_more_05 = 1 - norm.cdf(0.5, mu, sigma)

print(f"Вероятность ошибки меньше 0.4: {p_less_04:.4f}")
print(f"Вероятность ошибки больше 0.5: {p_more_05:.4f}")
```

# **Задача 12**
X нормально распределена с параметрами
a и
b. Найти вероятность, что
X попадёт в интервал
(a+c;a+2c).

```python
from scipy.stats import norm

# Заданные параметры
a, b, c = 10, 2, 3  # Здесь можно подставить нужные значения

# Вычисление вероятности
p_a_c = norm.cdf(a + 2 * c, loc=a, scale=b) - norm.cdf(a + c, loc=a, scale=b)

print(f"Вероятность (a+c < X < a+2c): {p_a_c:.4f}")
```

# **Задача 13**
Найти вероятность, что
X примет значение в интервале
(−2a−c;a+c).

```python
from scipy.stats import norm

# Заданные параметры (замени на нужные значения)
a, b, c = 10, 2, 3

# Вычисление вероятности
p_interval = norm.cdf(a + c, loc=a, scale=b) - norm.cdf(-2 * a - c, loc=a, scale=b)

print(f"P(-2a-c < X < a+c): {p_interval:.4f}")
```

# **Задача 25**
Среднее число клиентов в минуту — 2. Найти вероятность прихода за 4 минуты:

а) ровно 3 клиента

б) менее 3

в) не менее 3

Используем распределение Пуассона, так как поток клиентов — поток Пуассона.

```python
from scipy.stats import poisson

lambda_ = 2 * 4  # 2 клиента в минуту * 4 минуты

p_3 = poisson.pmf(3, lambda_)
p_less_3 = poisson.cdf(2, lambda_)
p_at_least_3 = 1 - p_less_3

print(f"P(3 клиента): {p_3:.4f}")
print(f"P(< 3 клиентов): {p_less_3:.4f}")
print(f"P(≥ 3 клиентов): {p_at_least_3:.4f}")
```
